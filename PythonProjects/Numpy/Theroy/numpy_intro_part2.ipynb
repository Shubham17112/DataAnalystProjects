```python
# generating array

import numpy as np

random_array = np.random.randint(100,size=5)  # 1D array 

print(random_array) # [81 96 25  4 65] shpae (3,)


# generating 2D array

import numpy as np

random_array = np.random.randint(100,size=(2,3))  # 2D array 

print(random_array)
```

```python
# Generate Random Number From Array  
# choise function 
import numpy as np
array = np.array([81 ,96 ,25 , 4 ,65])

x = np.random.choice(array) # will choose numenr from exiting array
print(x)  #81


# now u can add shape also in choice 

x = np.random.choice(array,size = (2,8))

print(x)
```

```python
# Normal Distribution  ( PDF ) 
# This is for continous random variable 
# - > alos know as gausian distributin where u will see standar devaitin , mean and varience
# Mean : all your data average value kisi eek point par lie karega thant called mean *loc
# Stand deviation : from mean how deviate your data  *scale 

# variance is square of standard deviation  
from numpy import random

x = random.normal(size=(2, 3))  # here standar deviatin is 1 and means also 1


print(x)
```

```python
# Visualization of Normal Distribution

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

x = np.random.normal(size=(10000))

print(x)  # u getting that array in which standar deviaiton is 1 and mean is 1 

sns.displot(x,kind="kde")
```

```python
# Binomial Distribution
# it's for discreate

# Three Parameter: 
#1st: number of trials, 2nd what is probability of sucucess ,3rd size of an array

# let's suppose u have onin and probabiltiy (H ) = 0.5 and u called it a sucess , u do 5 toos ever time and size of array is 10

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

array = np.random.binomial(p=0.5,n=5,size=10)

print(array)   # [2 2 3 2 3 3 4 3 3 3]  here 2 head got out of 5 trial  then again 2 head then 3 head so on .,.
```

```python
# visulization of binomial distribution 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

array = np.random.binomial(p=0.5,n=5,size=10)

sns.displot(array)
```
